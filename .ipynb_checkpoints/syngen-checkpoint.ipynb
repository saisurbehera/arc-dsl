```python
from glob import glob
import json 
import numpy as np

def read_all_data(file_loc):
    with open(file_loc , "r") as f:
        return json.loads(f.read())
    
all_files_loc = glob("data/training/*")
training_json = [read_all_data(i) for i in all_files_loc ]
```

```python
training_json = training_json[:len(all_solutions)*9//10]
```

```python
Counter([ all([ np.array(k).shape<(10,10) for j in i["train"] for k in j.values()]) for i in training_json ])
```

```python
all_possible_matrices_train = [tuple([tuple(row) for row in k]) 
                               for each in training_json  
                               for j in each["train"] if type(j) ==dict for k in j.values()]
```

```python
def do_work(i):
    args = random.choice(all_possible_matrices_train)
    function_name = random.choice(all_solutions)
    try:
        result = globals()[function_name](args)
        return (function_name,args)
    except:
        return None
```
